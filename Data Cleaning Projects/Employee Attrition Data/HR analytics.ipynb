# 2 Basic analysis 

```python
df_HR.head()
```

```python
df_HR.tail()
```

```python
x = ((df_HR['BusinessTravel'].value_counts()/df_HR['BusinessTravel'].value_counts().sum())*100).sort_values(ascending = False)
y = df_HR['BusinessTravel'].value_counts()

m_data = pd.concat([y, x], keys = ['Total', 'Percent'], axis = 1)
m_data
```

```python
a = ((df_HR['Department'].value_counts()/df_HR['Department'].value_counts().sum())*100).sort_values(ascending = False)
b = df_HR['Department'].value_counts()

m_data1 = pd.concat([b,a], keys = ['Total', 'Percent'], axis = 1)
m_data1
```

```python
c = ((df_HR['EducationField'].value_counts()/df_HR['EducationField'].value_counts().sum())*100).sort_values(ascending = False)
d = df_HR['EducationField'].value_counts()

m_data2 = pd.concat([d,c], keys = ['Total', 'Percent'], axis = 1)
m_data2
```
