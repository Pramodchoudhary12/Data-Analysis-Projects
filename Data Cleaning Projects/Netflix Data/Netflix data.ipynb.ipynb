# 2. Data Cleaning

1. Removing Unnessesary Null Values

```python
df_titles.dropna(subset = ['rating', 'date_added', 'duration'], inplace = True)
```

```python
df_titles.isnull().sum()
```

```python
df_titles = df_titles.reset_index()
```

2. Changing data type of 'date_added' column to DateTime

```python
df_titles['date_added'] = pd.to_datetime(df_titles['date_added'])
```

```python
df_titles
```

3. Finding Duplicates

```python
df_titles[df_titles.duplicated()]
```

4. Removing index and setting 'show_id' as Index

```python
df_titles.set_index('show_id')
```

```python
df_titles.drop('index', axis = 1, inplace = True)
```

```python
df_titles.set_index('show_id', inplace = True)
```

```python
df_titles
```

5. Removing unclosed colon " ' ".

```python
df_titles[df_titles['description'].str.contains("'")].count()
```

```python
df_titles['description'] = df_titles['description'].replace("'" ," " , inplace = True)
```
